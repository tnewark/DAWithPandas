# Filtering with the where() Method

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
mask = bond["Actor"] == "Sean Connery"
```

```python
bond.where(mask) # all rows returned... no match nulls it out
```

```python
mask = bond["Box Office" ] > 800
bond.where(mask)
```

```python
bond.where ((bond["Actor"] == "Sean Connery") & (bond["Box Office"] > 800))
```

# The .apply() Method with Row Values

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
def good_movie(row):
    actor = row[1]
    budget = row[4]
    if actor == "Pierce Brosnan": 
        return "The Best"
    elif actor == "Roger Moore" and budget > 40:
        return "Enjoyable"
    else:
        return "I don't know"
```

```python
bond.apply(good_movie, axis="columns")
```
